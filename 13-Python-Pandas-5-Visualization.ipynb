# Visualization with Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv("./datasets/iris.csv")
df
```

```python
df['variety'].value_counts()
```

# Pie

```python
df['variety'].value_counts().plot.pie()
```

```python
df['variety'].value_counts().plot.pie(colors=['g','r','w'],
                                     autopct="%.1f", fontsize=15,
                                     figsize=(4,4));
```

```python
# Defining colors for the pie chart 
colors = ['pink', 'silver', 'steelblue'] 
  
# Define the ratio of gap of each fragment in a tuple 
explode = (0.00, 0.00, 0.2) 

# Plotting the pie chart and rotating the pie chart by 60 degrees 
df['variety'].value_counts().plot.pie(colors=colors,
                                     autopct="%.1f", fontsize=15,
                                     figsize=(5,5), explode=explode, shadow=True,
                                     startangle=60);
```

```python
df = pd.DataFrame({'mass': [0.330, 4.87 , 5.97],
                    'radius': [2439.7, 6051.8, 6378.1]},
                   index=['Mercury', 'Venus', 'Earth'])
plot = df.plot.pie(y='mass', figsize=(5, 5))
```

```python
plot = df.plot.pie(subplots=True, figsize=(10, 5))
```

# Scatter

```python
df.plot.scatter(x='petal.length', y='petal.width',
               figsize=(10,8));
```

```python
df.variety.replace({'Setosa':0,'Versicolor':1,'Virginica':2},
                  inplace=True)
```

```python
df.plot.scatter(x='petal.length', y='petal.width',
               c='variety',colormap='cividis',figsize=(10,8),
               s=90);
```

# Hexbin

```python
np1 = np.array([2,5,6,2,7,2,5])
np2 = np.array([2,4,7,2,6,2,4])
df_2 = pd.DataFrame({'x': np1, 'y': np2})
df_2
```

```python
df_2.plot.hexbin(x='x',y='y', gridsize=10,
                figsize=(12,8));
```

```python
df.plot.hexbin(x='sepal.length',y='sepal.width', gridsize=10,
                figsize=(8,5), title="IRIS dataset");
```

```python
x = np.linspace(0,10,6)
y = np.power(2, x)
df_test = pd.DataFrame(y, columns=['4^x'])
df_test
```

```python
df_test.plot.line(figsize=(8,5),title="4^x");
```

```python
x = np.linspace(0,2*np.pi,1000)
y = np.sin(x)
df_test = pd.DataFrame(y, columns=['Sin(x)'])
df_test.plot.line(figsize=(8,5), title="Sin(X)");
```
