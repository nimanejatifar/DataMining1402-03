```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.DataFrame({
    'Day': np.linspace(1,20,20),
    'income': np.random.randint(2500,4000,20),
    'outcome': np.random.randint(1500,3500,20)

})
df1['net profit'] = df1['income'] - df1['outcome']
df1['total revenue'] = np.cumsum(df1['net profit'])
df1
```

```python
plt.figure(figsize=(10,6))
plt.plot(df1['Day'], df1['income'], label="INCOME", color='g')
plt.plot(df1['Day'], df1['outcome'], label="OUTCOME", color='r')

plt.grid(color='b', linewidth=1, linestyle= '--')
plt.fill_between(df1['Day'], df1['outcome'], color='r', alpha=0.4)
plt.legend()
plt.show()
```

```python
x = df['smoking_status'].unique()
y = df['smoking_status'].value_counts(sort = False)

plt.figure(figsize=(7,4))
plt.bar(x,y)
plt.show()
```

```python
x = df['work_type'].unique()
y = df['work_type'].value_counts(sort = False)

plt.figure(figsize=(7,4))
plt.bar(x,y, width=0.5 , fill=False, hatch='o')
plt.xlabel('work type')
plt.ylabel('count')
plt.title('Work Type', color = 'r')
plt.show()
```

```python
x = df['work_type'].unique()
y = df['work_type'].value_counts(sort = False)

plt.figure(figsize=(7,4))
plt.bar(x,y, width=0.5 , color='y',fill=True, hatch='/')
plt.xlabel('work type')
plt.ylabel('count')
plt.title('Work Type', color = 'r')
plt.grid(linestyle='--', axis='y', alpha=0.6)
plt.show()
```

```python
plt.pie(df['smoking_status'].value_counts(), 
        labels=["never smoked", "Unknown", "formerly smoked", "smokes"])
plt.show()
```

```python
labels=["never smoked", "Unknown", "formerly smoked", "smokes"]
x = df['smoking_status'].value_counts()
plt.figure(figsize=(5,5))
plt.pie(x, autopct="%.1f")
plt.title("Smoking Status")
plt.legend(labels, loc="best")
plt.show()
```

```python
labels=["never smoked", "Unknown", "formerly smoked", "smokes"]
x = df['smoking_status'].value_counts()
plt.figure(figsize=(5,5))
plt.pie(x, autopct="%.2f" , explode=[0,0,0,0.20], shadow=True)
plt.title("Smoking Status", fontsize= 20)
plt.legend(labels, loc=(0,0.3))
plt.show()
```
