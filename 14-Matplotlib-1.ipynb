```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 2*np.pi, 0.01)
y = np.sin(x)
plt.plot(x, y, color='green')
plt.show()
```

```python
df1 = pd.DataFrame({
    'Day': np.linspace(1,20,20),
    'income': np.random.randint(2000,4000,20),
    'outcome': np.random.randint(1500,3500,20)

})
df1
```

```python
df1['net profit'] = df1['income'] - df1['outcome']
df1
```

```python
df1['total revenue'] = np.cumsum(df1['net profit'])
df1
```

```python
plt.plot(df1['Day'], df1['net profit']);
```

```python
plt.figure(figsize=(8,6))
plt.plot(df1['Day'], df1['total revenue'],linestyle='--');
```

```python
plt.figure(figsize=(5,4))
plt.xlabel('Days')
plt.ylabel('Total Revenue')
plt.xlim(1,20)
plt.title('TOTAL')
plt.plot(df1['Day'], df1['total revenue'], color='k',linestyle='--', marker='v',
        markersize=10, markerfacecolor='yellow', label='revenue');
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,6))

plt.subplot(121)
plt.xlabel('Days')
plt.ylabel('Total Revenue')
plt.xlim(1,20)
plt.title('TOTAL')
plt.plot(df1['Day'], df1['total revenue'], color='k',linestyle='--', marker='v',
        markersize=10, markerfacecolor='yellow', label='revenue');

plt.subplot(122)
plt.xlabel('Days')
plt.ylabel('net profit')
plt.xlim(1,20)
plt.title('PROFIT')
plt.plot(df1['Day'], df1['net profit'], color='b',linestyle='-', marker='o',
        markersize=10, markerfacecolor='red', label='revenue');
plt.grid()
```
